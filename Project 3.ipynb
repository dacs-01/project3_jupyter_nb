```python
import pandas as pd
import numpy as np
df = pd.read_csv("train.csv")
df
```

```python
df.isna().sum()[df.isna().sum() > 0]
```

```python
nulls = df.isna().sum()[df.isna().sum() > 81]
df.drop(columns = nulls.index, axis=1, inplace=True)
```

```python
df.isna().sum()[df.isna().sum() > 0]
```

```python
df.dropna(inplace=True)
df
```

```python
df.dtypes
```

```python
from sklearn import preprocessing 
lab_en = preprocessing.LabelEncoder()
```

```python
strings = df.select_dtypes(include='object')
```

```python
strings = strings.apply(lab_en.fit_transform)
```

```python
strings.dtypes
```

```python
df_other_ints = df.select_dtypes(exclude=['object'])
```

```python
new_df = pd.concat([df_other_ints, strings], axis=1)
new_df.info(verbose=True)
```

```python
pd.set_option('display.max_rows', 80)
new_df.dtypes
```

```python
new_df.drop(columns='Id', axis=1)
new_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

X = new_df.drop(columns = ['SalePrice'])
y = new_df['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
lin_reg = LinearRegression()
lin_reg = lin_reg.fit(X_train, y_train)
```

```python
lin_reg.score(X_test, y_test)
```

```python
y_pred = lin_reg.predict(X_test)
```

```python
!pip install statsmodels
```

```python
import statsmodels.api as sma
```

```python
X2 = sma.add_constant(X_train)
est = sma.OLS(y_train, X2)
est2 = est.fit()
```

```python
est2.summary()
```

```python
from sklearn import metrics
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = metrics.mean_squared_error(y_test, y_pred,squared = False)

print('Mean Absolute Error:', mae)
print('Mean Square Error:', mse)
print('Root Mean Square Error:', rmse)
```
